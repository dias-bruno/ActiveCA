# Reading `ActiveCA` R Data Files and converting them to Pandas DataFrames

Importing Libraries:

```python
import pyreadr
import os
from pathlib import Path
```

Converting R Data Files into Pandas DataFrames:

```python
rmd_path = Path.cwd().parent / 'data'
```

```python
for file in os.listdir(rmd_path):
    file_name = os.path.splitext(file)[0]
    print("Creating a pandas data frame with the same name for the file: ", file_name)
    globals()[file_name] = pyreadr.read_r(os.path.join(rmd_path, file))[file_name]
```
